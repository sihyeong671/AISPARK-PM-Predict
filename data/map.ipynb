```python
import folium
import pandas as pd
```

```python
df_pm = pd.read_csv("../dataset/META/pmmap.csv")
df_aws = pd.read_csv("../dataset/META/awsmap.csv")
```

```python
df_pm.head()
```

```python
df_aws.head()
```

```python
print((df_pm["Latitude"].mean() + df_aws["Latitude"].mean()) / 2)
print((df_pm["Longitude"].mean() + df_aws["Longitude"].mean()) / 2)
```

```python
M = folium.Map(location=[36.549448382352935, 126.90016399313724], zoom_control=False, min_zoom=9, max_zoom=9, zoom_start=9)
for idx, pm in df_pm.iterrows():
  latitude, longitude = pm.Latitude, pm.Longitude
  html = f"<b>{pm.Location}<br>{pm.Description}</b>"
  iframe = folium.IFrame(html)
  popup = folium.Popup(iframe, min_width=200, max_width=300)
  folium.Marker([latitude, longitude], popup=popup).add_to(M)

for idx, pm in df_aws.iterrows():
  latitude, longitude = pm.Latitude, pm.Longitude
  html = f"<b>{pm.Location}<br>{pm.Description}</b>"
  iframe = folium.IFrame(html)
  popup = folium.Popup(iframe, min_width=200, max_width=300)
  folium.Marker([latitude, longitude], popup=popup, icon=folium.Icon(color='red')).add_to(M)

M
```
